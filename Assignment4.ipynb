```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import matplotlib.ticker as ticker
```

```python
df10 = pd.DataFrame()
years = np.arange(2007, 2020, 1)
years
```

```python
for year in years:
    df = pd.read_excel('Religiosidad Enero {}.xls'.format(year), header=6, index_col=0)
    df = df.iloc[0:5, 0:2]
    df.index = df.index.str.strip()
    df = df.rename(index={'Católico': 'Católico/a', 'Ateo': 'Ateo/a'})
    df10[year] = df['%/Total']
    
df10
```

```python
df10 = df10.transpose()
df10.columns = ['Catholic\n(Católico/a)',
                'Other religion\n(Creyente de otra religión)',
                'Non-religious\n(No creyente)',
                'Atheist\n(Ateo/a)',
                'No response']
df10.drop(['No response'], inplace=True, axis=1)
df10
```

```python
df10.plot(figsize=(15, 7), lw=3, ylim=(0, 100))

plt.title('How have religious beliefs in Spain changed in recent years?\n(¿Cómo han cambiado las creencias religiosas en España en los últimos años?)', fontsize='xx-large')
plt.ylabel('% of Survey Respondents\n(% de encuestados)', fontsize='large')
plt.legend(fontsize='large')

plt.grid()

plt.xticks(np.arange(min(years), max(years)+1, 1))

percentage = np.arange(0, 101, 1)
plt.yticks(np.arange(min(percentage), max(percentage)+1, 10))

plt.figtext(0.685, 0.05, "Source: Centro de Investigaciones Sociológicas")

plt.show()
```
